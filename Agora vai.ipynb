```python
import os
import pandas as pd
from scipy.spatial.distance import cdist, pdist, squareform
from sklearn.preprocessing import StandardScaler, normalize
```

```python
def tagnho(lista):
    if lista == 'No tags yet':
        return lista
    tag = eval(lista)
    dic = {}
    for tag, path in tag:
        dic[tag] = path
    return dic
```

```python
df_novo = pd.concat([pd.read_csv(f'data/dataset/{csv}') 
                                 for csv in os.listdir('data/dataset') 
                                    if 'sort=updated' in csv],
                    ignore_index=True).drop('Unnamed: 0', axis=1).dropna(how='all')

df_ref = pd.concat([pd.read_csv(f'data/dataset/{csv}')
                                for csv in os.listdir('data/dataset')
                                   if 'sort=hotness' in csv], 
                  ignore_index=True).drop('Unnamed: 0', axis=1).dropna(how='all')

df_ref.tags = df_ref.tags.apply(tagnho)
```

```python
df_hot = df_ref.select_dtypes(exclude=['object'])
drap = df_hot.columns
df_9 = df_novo[drap]
```
