```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
dfs = {}
for csv in os.listdir('data/tabs'):
    if 'Dataset' in csv or 'Tags.csv' == csv or 'Users' in csv:
        dfs[csv.split('.')[0]] = pd.read_csv(f'data/tabs/{csv}')
dfs.keys()
```

```python
for x in dfs.keys():
    print('\n\nTabela:',x, '\n')
    display(dfs[x])
```

```python
# tabela DatasetVersions

dfvers = dfs['DatasetVersions']
dfvers = dfvers.drop(['TotalCompressedBytes', 
                      'TotalUncompressedBytes'], axis=1)

dfvers[['VersionNumber', 
        'TotalCompressedBytes', 
        'TotalUncompressedBytes']] = dfvers['VersionNumber'].fillna(0)
dfvers.VersionNumber = dfvers.VersionNumber.astype('int64')

dfvers[['Description', 
        'VersionNotes', 
        'Subtitle']] = dfvers[['Description', 
                               'VersionNotes', 
                               'Subtitle']].fillna('not text here')    
```

```python
# tabela Datasets

# df =  dfs['Datasets']
df = dfs['Datasets'].merge(dfs['DatasetTags'], 
                               left_on='Id', 
                               right_on='DatasetId', 
                               suffixes=('', '_data_tag'))

df = df.merge(dfs['Tags'], 
                      left_on='TagId', 
                      right_on='Id', 
                      suffixes=('', '_tag'))

df = df.drop(['Type', 'ForumId', 'Id_data_tag',
                  'DatasetId', 'OwnerOrganizationId', 
                  'CurrentDatasetVersionId', 'Id_tag', 
                  'Description', 'CompetitionCount', 'OwnerUserId',
                  'KernelCount', 'ParentTagId', 'CurrentDatasourceVersionId'], axis=1)

df.CreationDate = pd.to_datetime(df.CreationDate)
df.LastActivityDate = pd.to_datetime(df.LastActivityDate)
# filtrona = (df.CurrentDatasetVersionId.isna()) | (df.CurrentDatasourceVersionId.isna())
# df.loc[filtrona] = df.fillna(0)


display(df.head(2))





# merge Datasets , DatasetTag, Tags, DatasetVotes, Users
# analise das atividades relcionadas ao datasets durante o tempo, 
# relacionar colunas com maior relacionamento com a popularidade(TotalViews, TotalDownloads, TotalVotes, TotalKernels)
# analizar tags
# diferenca entre as interaÃ§oes de cada lvl de usuarios
```
